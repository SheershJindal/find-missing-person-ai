```python
from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.model_selection import train_test_split
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv('/content/drug200.csv')
df.head()
```

```python
l1 = LabelEncoder()
l2 = LabelEncoder()
l3 = LabelEncoder()
df['Sex'] = l1.fit_transform(df['Sex'])
df['BP'] = l2.fit_transform(df['BP'])
df['Cholesterol'] = l3.fit_transform(df['Cholesterol'])
df.head()
```

```python
x = df.iloc[:,:5].values
x
```

```python
y = pd.get_dummies(df.iloc[:,5:]).values
y
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 24)
```

```python
# xtrain.space, xtest.space
```

```python
# ytrain.space, ytest.space
```

```python
model = Sequential()
model.add(Dense(5, activation = 'relu'))
model.add(Dense(36, activation = 'relu'))
model.add(Dense(24, activation = 'relu'))
model.add(Dense(18, activation = 'relu'))
model.add(Dense(5, activation = 'softmax'))
```

```python
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(xtrain, ytrain, batch_size = 4, epochs = 50, validation_data = (xtest, ytest))
```
